```python
import matplotlib.pyplot as plt
```

```python
# create a figure and axis
fig, ax = plt.subplots()

# scatter the sepal_length againts the sepal_width
ax.scatter(Food['briefing_time'], Food['cook_time'])
# set a title and labels
ax.set_title('Food Dataset')
ax.set_xlabel('prep_time')
ax.set_ylabel('cook_time')
```

```python
# create color dictionary
colors = {'vegetarian':'g', 'non vegetarian':'r'}
# create a figure and axis
fig, ax = plt.subplots()
# plot each data-point
for i in range(len(Food['prep_time'])):
    ax.scatter(Food['briefing_time'][i], Food['cook_time'][i],color=colors[Food['diet'][i]])
# set a title and labels
ax.set_title('Food Dataset')
ax.set_xlabel('briefing_time')
ax.set_ylabel('cook_time')
```

```python
# get columns to plot 
columns = Food.columns.drop(['diet', 'name', 'ingredients', 'briefing_time', 'cook_time', 'flavor_profile', 'course', 'state', 'region'])
# create x data
x_data = range(0, Food.shape[0])
# create figure and axis
fig, ax = plt.subplots()
# plot each column
for column in columns:
    ax.plot(x_data, Food[column], label=column)
# set title and legend
ax.set_title('Food Dataset')
ax.legend()
```

```python
# create figure and axis
fig, ax = plt.subplots()
# plot histogram
ax.hist(Food['prep_time'])
# set title and labels
ax.set_title('Food Scores')
ax.set_xlabel('prep_time')
ax.set_ylabel('cook_time')
```

```python
# create a figure and axis 
fig, ax = plt.subplots() 
# count the occurrence of each class 
data = Food['briefing_time'].value_counts() 
# get x and y data 
prep_time = data.index 
frequency = data.values 
# create bar chart 
ax.bar(prep_time, frequency) 
# set title and labels 
ax.set_title('Food Scores') 
ax.set_xlabel('briefing_time') 
ax.set_ylabel('Frequency')
```

```python
import numpy as np



# get correlation matrix

corr = Food.corr()

fig, ax = plt.subplots()

# create heatmap

im = ax.imshow(corr.values)



# set labels

ax.set_xticks(np.arange(len(corr.columns)))

ax.set_yticks(np.arange(len(corr.columns)))

ax.set_xticklabels(corr.columns)

ax.set_yticklabels(corr.columns)



# Rotate the tick labels and set their alignment.

plt.setp(ax.get_xticklabels(), rotation=45, ha="right",

         rotation_mode="anchor")
```
